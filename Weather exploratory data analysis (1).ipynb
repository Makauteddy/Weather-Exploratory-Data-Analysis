### 3.3 Correlation matrices and heatmaps to identify relationships between different weather patterns

#### 3.3.1 Correlation matrices

```python
# Select only numerical columns
numerical_data = data.select_dtypes(include=['float64', 'int64'])

# Calculate the Spearman correlation matrix and round to 2 decimal places
correlation_matrix = numerical_data.corr(method='pearson').round(2)

# Style the correlation matrix with a uniform green background color
styled_matrix = correlation_matrix.style.background_gradient(axis=None, cmap='Greens')

# Display the styled correlation matrix
styled_matrix
```

```python
# Select only numerical columns
numerical_data = data.select_dtypes(include=['float64', 'int64'])

# Calculate the Spearman correlation matrix
correlation_matrix = numerical_data.corr(method='spearman').round(2)

# Create a mask for the upper triangle
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

# Create a figure with a size of 8x8
plt.figure(figsize=(8, 8))

# Create a heatmap with color gradient
sns.heatmap(correlation_matrix, 
            annot=True,                 # Annotate with correlation values
            cmap='Greens',              # Use green colormap
            mask=mask,                  # Mask the upper triangle
            vmin=-1, vmax=1,            # Set color range
            center=0,                   # Center the colormap at 0
            linewidths=0.5,            # Linewidths between cells
            linecolor='black',         # Line color between cells
            cbar_kws={'label': 'Correlation Value'})  # Colorbar label

# Add labels and title
plt.title('Spearman Correlation Matrix')  # Plot title

# Display the plot
plt.tight_layout()  # Adjust layout to fit labels
plt.show()
```
